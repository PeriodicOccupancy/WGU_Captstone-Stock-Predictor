```python
# Run this block first, Shift + Enter
import sklearn
import numpy as np
import pandas as pd
import pandas_datareader as pdr
import yfinance as yf
import matplotlib.pyplot as plt


stock_symbols = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'META']
```

```python
#Run this block 2nd.

#block to create csv files of stock data from yfinance
#stock_symbols = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'META']
start_date = "2013-01-01"
end_date = "2023-01-01"

stock_data = {}
for symbol in stock_symbols:
    stock_data[symbol] = yf.download(symbol, start=start_date, end=end_date)  
    stock_data[symbol].to_csv(f"{symbol}_stock_data.csv")

```

```python
# Run this block 3rd

#block for data analysis
#stock_symbols = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'META']


for symbol in stock_symbols:
    file = f"{symbol}_stock_data.csv"
    data = pd.read_csv(file)

    print(f"\n\n\n{symbol}\n")
    print(f"Head\n{data.head()}\n")
    print(f"Info\n")
    display(data.info())

    print(f"\nDescribe\n{data.describe()}\n")
```

```python
# Run this block 4th

#block for feature engineering
#stock_symbols = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'META']

for symbol in stock_symbols:
    file = f"{symbol}_stock_data.csv"
    data = pd.read_csv(file)
    data['Date'] = pd.to_datetime(data['Date'])
    data['Year'] = data['Date'].dt.year
    data['Month'] = data['Date'].dt.month
    data['Day'] = data['Date'].dt.day
    data['DayOfWeek'] = data['Date'].dt.dayofweek
    
    data['DailyRange'] = data['High'] - data['Low']
    data['MA5'] = data['Close'].rolling(window=5).mean()
    data['MA20'] = data['Close'].rolling(window=20).mean()
    
    def calculate_rsi(data, window=14):
        delta = data['Close'].diff()
        gain = delta.where(delta > 0, 0)
        loss =-delta.where(delta < 0, 0)
        avg_gain = gain.rolling(window=window).mean()
        avg_loss = loss.rolling(window=window).mean()
        rs = avg_gain / avg_loss
        rsi = 100 - (100 / (1 + rs))
        return rsi
    
    data['RSI'] = calculate_rsi(data)
    
    forecast_range = 7
    for i in range(1, forecast_range + 1):
        data[f'Target_{i}'] = data['Close'].shift(-i)
    
    #data.dropna(subset=[f'Target_{i}' for i in range(1, forecast_range + 1)], how='all', inplace =True)
    data.drop(columns=['Date'], inplace=True)
    data.dropna(inplace=True)
    data.to_csv(f"{symbol}_updated.csv", index=False)
```

```python
# Run this block 5th 

from sklearn.model_selection import train_test_split

#Train Test Split
for symbol in stock_symbols:
    file = f"{symbol}_updated.csv"
    filedata = pd.read_csv(file)

    num_targets = 7
    target_columns = [f"Target_{i}" for i in range(1, num_targets + 1)]
    features = filedata.drop(columns=target_columns)
    targets = filedata[target_columns]

    x_train, x_test, y_train, y_test = train_test_split(features, targets, test_size=0.2, shuffle = False, random_state=42)

    x_train.to_csv(f'x_train_{symbol}.csv', index=False) #input training
    x_test.to_csv(f'x_test_{symbol}.csv', index=False) #features for testing

    y_train.to_csv(f'y_train_{symbol}.csv', index=False) #target training
    y_test.to_csv(f'y_test_{symbol}.csv', index=False) #target variables for testing
```

```python
# Run this block last
# This block constains the model and visualizations of the predictions

from sklearn.linear_model import LinearRegression


# This Variable can be changed to any of the following: 'AAPL', 'MSFT', 'GOOGL', 'AMZN', 'META'
symbol = "AAPL"

x_train = pd.read_csv(f'x_train_{symbol}.csv')
x_test = pd.read_csv(f"x_test_{symbol}.csv")

y_train = pd.read_csv(f"y_train_{symbol}.csv")
y_test = pd.read_csv(f"y_test_{symbol}.csv")


model = LinearRegression() #base model
model.fit(x_train, y_train) #train the model on the testing data

y_pred = model.predict(x_test) #make predictions

mse = sklearn.metrics.mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print(f"Root Mean Squared Error: {rmse}")
for i in range (y_test.shape[1]):
    plt.figure(figsize=(10, 6))
    plt.plot(y_test.iloc[:, i], label=f'Actual Target {i+1}')
    plt.plot(y_pred[:, i], label=f'Predicted Target {i+1}')
    plt.legend()
    plt.title(f"Actual vs. Predicted Stock Prices for Target {i+1}")
    plt.xlabel("Data Points")
    plt.ylabel("Stock Price")
    plt.show()
```
